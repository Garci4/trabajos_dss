```python
import numpy as np
import pandas as pd
from sklearn import tree

input_file = "winequality-red.csv"
df = pd.read_csv(input_file, header = 0)

df.head()
```

```python
df = pd.read_csv('winequality-red.csv', delimiter=';')

df.head()
```

```python
features = list(df.columns[:11])
features
```

```python
df.drop_duplicates('quality')
```

Se define el nivel de calidad bueno >=5 y malo <5

```python
d = {1: 0, 2: 0, 3: 0, 4: 0, 5: 1, 6: 1, 7: 1, 8: 1}
df['quality'] = df['quality'].map(d)

df.drop_duplicates('quality')
```

```python
y = df['quality'] # resultados esperados o labels
x = df[features] # datos de entrenamiento
clf = tree.DecisionTreeClassifier()
clg = clf.fit(x,y)

```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(25,25))
_ = tree.plot_tree(clg, 
                   feature_names=features,  
                   class_names=["bad","good"],
                   filled=True)
```
