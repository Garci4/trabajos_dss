```python
import pandas as pd
```

```python
df = pd.read_csv("covid-cases-raw.csv")
```

```python
dp = pd.read_csv("departamentos.csv")
dp.head()
```

```python
dp.drop_duplicates()
```

```python
pv = pd.read_json("provincias.json")
pv.head()
```

```python
pv.rename(columns = {'id':'provincia_id'}, inplace = True)
```

```python
pv.rename(columns = {'name':'nombre_prov'}, inplace = True)
pv.rename(columns = {'pop_dens':'dens_pob_prov'}, inplace = True)
pv.rename(columns = {'total_pop':'poblacion_prov'}, inplace = True)
pv.rename(columns = {'country_percentage':'porcentaje_pob_pais'}, inplace = True)
```

```python
pv.head()
```

```python
df.rename(columns = {'carga_provincia_id':'provincia_id'}, inplace = True)
```

```python
df.head()
```

```python
merge = pd.merge(df, pv, on="provincia_id", how='outer')
```

```python
merge.head()
```

```python
dp.rename(columns = {'id':'residencia_departamento_id'}, inplace = True)
dp.head()
```

```python
#se vuelve a cambiar el nombre por comodidad
dp.rename(columns = {'departamento_id':'residencia_departamento_id'}, inplace = True)
dp.head()
```

```python
_merge = pd.merge(merge, dp, on="residencia_departamento_id", how="outer")
```

```python
_merge.rename(columns = {'label':'unidad_medida_edad'}, inplace = True)
_merge.rename(columns = {'class_0008':'caso_covid'}, inplace = True)
_merge.rename(columns = {'nombre':'nombre_departamento'}, inplace = True)
```

```python
#Unificacion de tablas provincias y departamentos con el df inicial
_merge.head()
```
